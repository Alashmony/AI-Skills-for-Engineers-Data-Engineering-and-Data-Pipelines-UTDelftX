```python
# follow code will rename the column names and convert datetime columns

flights = (
    flights.rename(
        columns={
            "FILED OFF BLOCK TIME": "fobt",
            "FILED ARRIVAL TIME": "fat",
            "ACTUAL OFF BLOCK TIME": "aobt",
            "ACTUAL ARRIVAL TIME": "aat",
            "STATFOR Market Segment": "market",
            "ADEP": "origin",
            "ADES": "destination",
            "ADEP Latitude": "lat_orig",
            "ADEP Longitude": "lon_orig",
            "ADES Latitude": "lat_dest",
            "ADES Longitude": "lon_dest",
            "AC Type": "actype",
            "AC Operator": "operator",
            "AC Registration": "reg",
            "ICAO Flight Type": "flighttype",
            "ECTRL ID": "id",
            "Requested FL": "requestfl",
            "Actual Distance Flown (nm)": "distance",
        }
    )
    .assign(fobt=lambda d: pd.to_datetime(d.fobt, format="%d-%m-%Y %H:%M:%S", utc=True))
    .assign(aobt=lambda d: pd.to_datetime(d.aobt, format="%d-%m-%Y %H:%M:%S", utc=True))
    .assign(fat=lambda d: pd.to_datetime(d.fat, format="%d-%m-%Y %H:%M:%S", utc=True))
    .assign(aat=lambda d: pd.to_datetime(d.aat, format="%d-%m-%Y %H:%M:%S", utc=True))
)
```

```python
# function to calculate great circle distance between two locations
def calc_gcd(r):
    lat1 = np.radians(r.lat_orig)
    lon1 = np.radians(r.lon_orig)
    lat2 = np.radians(r.lat_dest)
    lon2 = np.radians(r.lon_dest)

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    d = c * 6371 * 0.539957  # nautical miles
    return d
```

## Query flights from June 15th, and then calculate the great circle distance with apply()

```python

flights.query('aobt.dt.day == 15').apply(calc_gcd, axis = 1)
```

### With one nest command: 
- add a new column named **gcd** (great circle distance) for all the flights
- then add another new column that calculates the different between **distance** and **gcd**, with column name **distance_diff**

```python
flights = flights.assign(gcd = flights.apply(calc_gcd, axis=1))
```

```python
flights.head(1)
```

```python
flights = flights.assign(distance_diff = lambda x: x.distance - x.gcd)
```

### Nesting operations

- Query flights originated from Paris Charles de Gaulle Airport (airport code: LFPG)
- Calculate arrival delays of all airports
- Group by destination
- Calculated the mean flight distance and average arrival delay at the destination airports
- Sow and sort results for destinations with average arrival delay large than 10 minutes with more than 5 flights

```python
flights.columns
```

```python
flights.query('origin == "LFPG"').assign(delay= lambda x : (x.fat-x.aat).dt.total_seconds() / 60).groupby('destination')\
    .agg({'distance': 'mean', 'delay':'mean','id':'count'})\
    .query('delay > 10 and id > 5').sort_values(by='delay', ascending=False).reset_index()
```
