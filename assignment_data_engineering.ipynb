```python
# function to calculate great circle distance between two locations
def calc_gcd(r):
    lat1 = np.radians(r.lat_orig)
    lon1 = np.radians(r.lon_orig)
    lat2 = np.radians(r.lat_dest)
    lon2 = np.radians(r.lon_dest)

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    d = c * 6371 * 0.539957  # nautical miles
    return d
```

### Query flights from June 15th, and then calculate the great circle distance with apply() (1pt)

```python
flights.query("aobt.dt.day == 15").apply(calc_gcd, axis=1)
```

### With one nest command: (2 pts)
- add a new column named **gcd** (great circle distance) for all the flights
- then add another new column that calculates the different between **distance** and **gcd**, with column name **distance_diff**

```python
flights = flights.assign(gcd=flights.apply(calc_gcd, axis=1)).assign(
    distance_diff=lambda x: x.distance - x.gcd
)
```

### Nesting operations (3 pts)

- Query flights originated from Paris Charles de Gaulle Airport (airport code: LFPG)
- Calculate arrival delays of all airports
- Group by destination
- Calculated the mean flight distance and average arrival delay at the destination airports
- Sow and sort results for destinations with average arrival delay large than 10 minutes with more than 5 flights

```python
flights.columns
```

```python
flights.query('origin == "LFPG"').assign(
    delay=lambda x: (x.fat - x.aat).dt.total_seconds() / 60
).groupby("destination").agg(
    {"distance": "mean", "delay": "mean", "id": "count"}
).query(
    "delay > 10 and id > 5"
).sort_values(
    by="delay", ascending=False
).reset_index()
```

### Using boxplot() to plot the arrival and departure delay statistics for each of the top 15 airport (1 pts)
 - remark: top airports according to number of total departures

```python
flights.columns
```

```python
top_15_dep_airports = (
    flights.groupby("origin")
    .agg({"id": "count"})
    .sort_values(by="id", ascending=False)
    .head(15)
)
# display(top_15_airports)

departure_airport = (
    flights.merge(top_15_dep_airports, how="inner", on="origin")
    .assign(delay=lambda d: (d.fobt - d.aobt).dt.total_seconds() / 60)
    .assign(airport_type="origin")
    .assign(flight_type="depart")
    .assign(airport=lambda x: x.origin)
)

top_15_dest_airports = (
    flights.groupby("destination")
    .agg({"id": "count"})
    .sort_values(by="id", ascending=False)
    .head(15)
)
# display(departure_airport)
arrival_airports = (
    flights.merge(top_15_dest_airports, how="inner", on="destination")
    .assign(delay=lambda d: (d.fat - d.aat).dt.total_seconds() / 60)
    .assign(airport_type="destination")
    .assign(flight_type="arrival")
    .assign(airport=lambda x: x.destination)
)
# display(arrival_airports)
airports = pd.concat([departure_airport, arrival_airports])
# display(airports)

fig, ax = plt.subplots(1, figsize=(15, 8))
sns.boxplot(data=airports, x="airport", y="delay", hue="flight_type", ax=ax)
ax.set_xlabel("airport")
ax.set_ylabel("hourly departing flights")
plt.xticks(rotation=45)
ax.set_ylim(top=1200)
plt.show()
```

### Use jointplot() to show the relationship between departure and arrival delays with 2D kde plot, you should:  (2 pts)   

 - randomly sample 10000 flights for the plot for better performance
 - limit showing delays between -60 and 60 minutes for better visualization
 

```python
data = (
    flights.sample(10000)
    .assign(departure_delay=lambda d: (d.fobt - d.aobt).dt.total_seconds() / 60)
    .assign(arrival_delay=lambda d: (d.fat - d.aat).dt.total_seconds() / 60)
    .query(
        "departure_delay < 60 and departure_delay > -60 and arrival_delay < 60 and arrival_delay > -60"
    )
)

sns.jointplot(data=data, x="departure_delay", y="arrival_delay", kind="kde", fill=True)
plt.show()
```
